# **2️⃣ plot() 메소드**
## 그래프를 만들어내는 메소드
## **matplotlib.pyplot.plot( \*args, scalex=True, scaley=True, data=None, **kwargs)**

### *args : 보통 x,y 값으로, array-like한 데이터가 들어감. 그 외 다른 것도 가능하나, 그건 하면서 파악.


```python
# 리스트 하나만 넣으면 y값으로 인식하고, 자동으로 x값은 정수 0,1,2,3,4 에 매핑됨.
plt.plot([1,2,3,4,5])
```

```python
# 리스트를 두 개 넣으면, [x],[y]로 인식함. 주의점은, x값과 y값은 같은 크기를 가져야 함.
plt.plot([1,2,3,4,5],[2,4,6,8,10])  #x값이 1,2,3,4,5 일 때, y값이 각각 2,4,6,8,10 인 점들을 연결한 그래프
plt.title("graph_01") #그래프의 타이틀
plt.show()
```

```python
# numpy array 를 투입할 수 있음
data = np.arange(1,10)

plt.plot(data)
plt.title("graph_01") #그래프의 타이틀
plt.show()
```

```python
# pandas df를 투입할 수 있음.
df = pd.DataFrame({'x':[1,2,3,4,5],'y':[2,4,6,8,10]}) #x칼럼 y칼럼을 만들어서

plt.plot(df['x'],df['y'])  #각 칼럼을 x,y 인자로 투입.
plt.title("graph_01") #그래프의 타이틀
plt.show()
```

```python
# 캔버스를 수정해보자.
plt.figure(figsize=(10,5)) #가로 10인치, 세로 5인치 직사각형 캔버스를 준비함(mpl default ppi = 100dpi, 72dpi 기준, 72px) 👉 x,y좌표에는 변화 없음에 유의!
plt.plot([1,2,3,4,5],[2,4,6,8,10])
plt.title("graph_01") #그래프의 타이틀
plt.show()
```

```python
# 캔버스 컬러를 바꿔보자.
plt.figure(figsize=(10,5),facecolor='gray') #facecolor 를 넣으면 figure 영역과 axes 영역을 구분해서 볼 수 있다.
plt.plot([1,2,3,4,5],[2,4,6,8,10])
plt.title("graph_01") #그래프의 타이틀
plt.show()
```

```python
# 스타일 종합
plt.plot([1,2,3,4,5],[2,4,6,8,10],
         color="dodgerblue",
         linewidth=2,
         linestyle="-.",
         marker="o",
         ms=7,
         mec="violet",
         mew=2,
         mfc="mediumslateblue")
plt.show()
```

## matplotlib.pyplot.xticks(ticks=None, labels=None, *, minor=False, **kwargs)
## ▪️ tics : 각 눈금들의 위치(position)를 담은 array-like한 것
## ▪️ labels : 각 눈금들의 이름(label)을 담은 array-like한 것

<br>

## plt.xticks(ticks = np.arange(start,stop,step)) 👉 x축 눈금 간격
## plt.yticks(ticks = np.arange(start,stop,step)) 👉 y축 눈금 간격
## plt.xticks(rotation='angle') 👉 x눈금 문자열 회전
## plt.yticks(rotation='angle') 👉 y눈금 문자열 회전

```python
# 그냥 plot을 만들면 default 값은 아래와 같음.
# plot의 범위는 x(0~9),y(0~9)인데 눈금은 0,2,4,6,8

x = np.arange(0,10)
y = np.arange(0,10)

plt.plot(x,y)

plt.show()
```

```python
# ticks를 plot의 x,y값과 동일하게 해주면,
# 중간을 건너뛰지 않은, plot : tics = 1:1 대응의 tics가 생김

x = np.arange(0,10)
y = np.arange(0,10)

plt.plot(x,y)

plt.xticks(ticks = x)
plt.yticks(ticks = y)

plt.show()
```

```python
# plot이 첨예해지면?

x = np.arange(0,100)
y = np.arange(0,100)

plt.plot(x,y)

plt.xticks(ticks = np.arange(0,100))
plt.yticks(ticks = np.arange(0,100))

plt.show()
```

```python
# plot이 첨예해지면 tics를 모두 표기하기보다, 간격을 두는게 나음.
# 그래서 tics는 plot의 (x,y) 개수와 1:1 대응이 아닌, 작은 값을 가질 수 있음.

x = np.arange(0,100)
y = np.arange(0,100)

plt.plot(x,y)

plt.xticks(ticks = np.arange(0,100,5)) #x축은 5간격 눈금
plt.yticks(ticks = np.arange(0,100,10)) #y축은 10간격 눈금

plt.show()
```

```python
# 애당초,'plot이 어디까지만 그려지는가' 의 관점으로 바라보면,
# tics는 plot의 (x,y) 개수 초과로 존재할 수 있음.

# 아래는 tics는 15x15 지만, plot은 10,10 까지만 그려진 그래프임.

x = np.arange(0,10)
y = np.arange(0,10)

plt.plot(x,y)

plt.xticks(ticks = np.arange(0,15,1))
plt.yticks(ticks = np.arange(0,15,1))

plt.show()
```

```python
# tics 는 가독성을 위해 회전시킬 수 있음

x = np.arange(0,100)
y = np.arange(0,100)

plt.plot(x,y)

plt.xticks(ticks = np.arange(0,100,3), rotation = 90) #x축은 3간격 눈금, 90도 회전
plt.yticks(ticks = np.arange(0,100,10)) #y축은 10간격 눈금

plt.show()
```

```python
# 각 tics 에는 label을 할당할 수 있고,
# tics : label = 1:1 로, 무조건 갯수가 일치해야 함.

x = np.arange(0,5)
y = np.arange(0,5)

plt.plot(x,y)

plt.xticks(ticks = x, labels=['a','b','c','d','e']) #x축은 3간격 눈금, 90도 회전
plt.yticks(ticks = y, labels=['A','B','C','D','F']) #y축은 1간격 눈금

plt.show()
```

# **5️⃣ subplots()** 메소드와 객체

## 여러개의 plot들을 그리는 메소드.

## **matplotlib.pyplot.subplots(nrows=1, ncols=1, *, sharex=False, sharey=False, squeeze=True, width_ratios=None, height_ratios=None, subplot_kw=None, gridspec_kw=None, **fig_kw)**


## **nrows** : 각 plot을 matrix에 속한 하나의 element 라고 봤을 때, 행(row) 수
## **ncols** : 각 plot을 matrix에 속한 하나의 element 라고 봤을 때, 열(column) 수
## * : plot을 matrix에 속한 하나의 element 라고 봤을 때, 그 **위치(Position)**
## (글 처럼, 좌 -> 우로, 상 -> 하로 순서가 매겨짐.)

<br>

## **return** : fig, ax (object)
## 👉figure 객체와 Axes 객체를 튜플로 리턴한다.

##✅ **subplots()의 반환값 (fig,ax)**
## fig = figure 객체 = 모든 subplots를 담고있는 전체 캔버스 하나
## ax = axes 객체 = 모든 subplots (사실상 여러개임)
## 튜플로 리턴함.

```python
#figure 객체만 만들어보기
f = plt.figure(figsize = (10,10))
print(type(f))
```

```python
#ax 객체만 만들어보기
a = plt.axes()
print(type(a))
```

```python
#plt.subplots()의 리턴값이 fig, ax 객체를 튜플로 패킹한 것이기에, 보통 아래와 같이 변수에 할당함.
fig, ax = plt.subplots()
print(type(fig),type(ax))
```

##✅ **nrows, ncols**
## **nrows** : 각 plot을 matrix에 속한 하나의 element 라고 봤을 때, 행(row) 수
## **ncols** : 각 plot을 matrix에 속한 하나의 element 라고 봤을 때, 열(column) 수

```python
#nrows, ncols
#10x5 inch 사이즈의 figure 안에
#1행 2열 (1x2 shape) 형태로 plot들을 배열
fig, ax = plt.subplots(nrows=1, ncols=2, figsize =(10,5))
```

```python
#nrows, ncols
#10x10 inch 사이즈의 figure 안에
#2행 2열 (2x2 shape) 형태로 plot들을 배열
fig, ax = plt.subplots(nrows=2, ncols=2, figsize =(10,5))
```

```python
# ax 는 numpy array임.
print(type(ax))
print(ax)
```

##✅ **여러 그래프 그리는 법 1 : subplots() 각 요소에 접근하기**
## ▪ 모든 plot들이 1차원 array 된 상태 = **ax[]** (1차원 인덱싱 접근)
## ▪ 모든 plot들이 2차원 array 된 상태 = **ax[][]** (2차원 인덱싱 접근)

```python
# 1차원 array
# 먼저 figurs 캔버스와 plots의 틀을 만들어주고
fig, ax = plt.subplots(nrows=1,ncols=3,figsize=(12,4))

# 표시할 데이터들을 준비하고
x_01 = np.arange(1,10)
x_02 = np.arange(10,19)
y_01 = np.random.randint(1,100,9)
y_02 = np.random.randint(1,100,9)

# 각 위치의 그래프를 인덱스로 접근하여 plot()으로 그려준다.
ax[0].plot(x_01,y_01, color='green',marker='o',linestyle='--')
ax[1].plot(x_02,y_02, color='red',marker='^',linestyle='-.')
ax[2].plot(x_01,y_02, color='blue',marker='*',linestyle=':')

#아래 그려진 그래프가 ax[0], ax[1], ax[2] 순임에 주목.
```

## matplotlib.pyplot.subplot(\*args, **kwargs)
## \*args -> 그래프가 그려질 subplots의 shape와 position (예: 2x2 array의 첫번째 위치 = 2,2,1)

```python
# 2차원 array
# 먼저 캔버스와 plot의 틀을 만들어주고
fig, ax = plt.subplots(nrows=2,ncols=2,figsize=(10,10))

# 표시할 데이터들을 준비하고
x_01 = np.arange(1,10)
x_02 = np.arange(10,19)
y_01 = np.random.randint(1,100,9)
y_02 = np.random.randint(1,100,9)

# subplot()메소드를 활용하여 각 그래프의 위치를 지정해줌
ax1 = plt.subplot(2,2,1)
ax1.plot(x_01,y_01, color='green',marker='o',linestyle='--')
ax1.set_title('2,2,1')

ax2 = plt.subplot(2,2,2)
ax2.plot(x_02,y_02, color='red',marker='^',linestyle='-.')
ax2.set_title('2,2,2')

ax3 = plt.subplot(2,2,3)
ax3.plot(x_01,y_02, color='blue',marker='*',linestyle=':')
ax3.set_title('2,2,3')

ax4 = plt.subplot(2,2,4)
ax4.plot(x_02,y_01, color='violet',marker='+',linestyle='-')
ax4.set_title('2,2,4')

```

##✅ **Bar 차트**
## 👉 데이터의 유형이 범주형(categorical)인 경우, 자주 쓰는 타입.
## 👉 X축에 범주형, 혹은 이산(discrete)값이 오고, Y축에는 연속(continuous)형 값이 와야 하는경우 주로 활용되는 차트
## matplotlib.pyplot.bar(x, height, width=0.8, bottom=None, *, align='center', data=None, **kwargs)

<br>

## ▪️ x : 표에서 x값들 (카테고리들)
## ▪️ height : 각 카테고리별 y값
## ▪️ width : bar의 폭
## ▪️ bottom : 어느 그래프를 아래에 둘지 지정. (누적 막대 그래프)
## ▪️ label : 막대그래프를 2세트 이상 출력시, 각 막대그래프 세트 의 이름

```python
#❗ 2개 이상의 비교군 bar 그래프 그리기.

# 그냥하면 각 bar 그래프의 position이 겹쳐서,
# 그래프가 앞뒤로 겹쳐버림 (🤨위아래로 겹친거 아님!!)

#데이터(학생 2명의 국영수사과 점수랜덤)
x = np.arange(5)
xlabel = ['Korean','English','Math','Science','Social']
y_1 = np.random.randint(50,100,5)
y_2 = np.random.randint(50,100,5)

#학생 2명의 그래프 표시. 여기서 label은 <각 막대그래프 세트> 의 이름
plt.bar(x, y_1 , label = 'student_A')
plt.bar(x, y_2 , label = 'student_B')

#axes의 tics, label을 지정
plt.xticks(x, xlabel)
plt.ylabel('score')
plt.legend(loc='upper right') # 막대그래프 세트의 이름이 부여된 경우, 범례 출력 가능.
plt.show
```

```python
#❗ 2개 이상의 비교군 bar 그래프 그리기.

# 그냥하면 각 bar 그래프의 position이 겹쳐서,
# 그래프가 앞뒤로 겹쳐버림 (🤨위아래로 겹친거 아님!!)
# 그래서 bar 그래프의 폭(width)를 지정해서, 손수 positioning을 해줘야 함 (귀찮게스리)


#데이터(학생 2명의 국영수사과 점수랜덤)
x = np.arange(5)
xlabel = ['Korean','English','Math','Science','Social']
y_1 = np.random.randint(50,100,5)
y_2 = np.random.randint(50,100,5)

# bar 그래프의 폭(width) 지정
bar_width = 0.3

#학생 2명의 그래프 표시
plt.bar(x, y_1 , width = bar_width,  label = 'student_A')
plt.bar(x + bar_width, y_2, width = bar_width,  label = 'student_B')
 #두 번째 그래프의 position은 첫 그래프의 position에서 bar_width 만큼 x축 양의 방향 이동해야 함.

#axes의 tics, label을 지정
plt.xticks(x + bar_width/2, xlabel)  # bar_width 의 절반만큼만 x축 양의 방향 이동하면 tics도 맞음
plt.ylabel('score')
plt.legend(loc='upper right')
plt.show
```

```python
#❗ 2개 이상의 비교군 bar 그래프 그리기.
# 누적 막대 그래프는 bottom kwargs를 사용하면 됨.

#데이터(학생 2명의 국영수사과 점수랜덤)
x = np.arange(5)
xlabel = ['Korean','English','Math','Science','Social']
y_1 = np.random.randint(50,100,5)
y_2 = np.random.randint(50,100,5)

# bar 그래프의 폭(width) 지정
bar_width = 0.3

#학생 2명의 그래프 표시
plt.bar(x, y_1, width = bar_width,  label = 'student_A', bottom= y_2)
# 어느 그래프가 bottom에 오게 할지 kwargs로 지정
# 자기 자신은 bottom에 오게 할 이유가 없음. 출력값만 2배로 복제될 뿐
plt.bar(x, y_2, width = bar_width,  label = 'student_B')

#axes의 tics, label을 지정
plt.xticks(x, xlabel)
plt.ylabel('score')
plt.legend(loc='upper right')
plt.show
```

```python
# bar의 컬러 및 스타일링도 가능. (스타일링은 공식문서 참고.)

#데이터(학생 2명의 국영수사과 점수랜덤)
x = np.arange(5)
xlabel = ['Korean','English','Math','Science','Social']
y_1 = np.random.randint(50,100,5)
y_2 = np.random.randint(50,100,5)

# bar 그래프의 폭(width) 지정
bar_width = 0.3

#학생 2명의 그래프 표시
plt.bar(x, y_1, width = bar_width,  label = 'student_A', bottom= y_2, color = '#6a25cc')
plt.bar(x, y_2, width = bar_width,  label = 'student_B', color = '#eb75e7')

#axes의 tics, label을 지정
plt.xticks(x, xlabel)
plt.ylabel('score')
plt.legend(loc='upper right')
plt.show
```

##✅ **Pie 차트**
## 👉 데이터의 유형이 범주형(categorical)이고, 상대 비중을 보고 싶을 때.
## matplotlib.pyplot.pie(x, explode=None, labels=None, colors=None, autopct=None, pctdistance=0.6, shadow=False, labeldistance=1.1, startangle=0, radius=1, counterclock=True, wedgeprops=None, textprops=None, center=(0, 0), frame=False, rotatelabels=False, *, normalize=True, hatch=None, data=None)

<br>

## ▪️ x : 표에서 각 파이조각(wedge)의 크기들
## ▪️ explode : 각 파이들의 돌출정도를 튜플로 별도 정의하여 적용함.
## ▪️ colors : 특정 색상 or 컬러팔레트를 적용할 수 있음  
## ▪️ kwargs : shadow, autopct 정도만 알아두자
## shadow : 그림자로 파이를 입체화
## autopct : auto percentage 자동 백분율 계산.

```python
#데이터(스포츠 웨어 브랜드 인플루언서 언급률)
brands = ['Nike','Gymshark','Adidas','Underarmour','others']
wedges = [20,12,8,5,55]


#axes의 tics, label을 지정
plt.pie(wedges, labels = brands)
plt.title('sportswear brands most mentioned')
plt.show
```

```python
#여러가지 후가공이 된 파이차트

import seaborn as sns  # 컬러 팔레트 사용을 위한 seaborn 라이브러리 임포트

ex_rate = (0.3,0,0,0,0) # 각 wedge들의 explode 돌출율을 위한 튜플값

palette = sns.color_palette('pastel',len(brands)) # pastel 팔레트에서 brand 수 만큼 색상 가져오기

plt.pie(wedges, explode = ex_rate, labels = brands, colors = palette, shadow=True, autopct='%0.1f%%')
#explode = 돌출율 적용, colors = 팔레트 적용, shadow = 그림자 적용, autopct = 자동 비중계산 적용 + 포매팅(소수 첫째자리수, '%'기호 추가)
plt.title('sportswear brands most mentioned')
plt.show
```

## matplotlib.pyplot.hist(x, bins=None, range=None, density=False, weights=None, cumulative=False, bottom=None, histtype='bar', align='mid', orientation='vertical', rwidth=None, log=False, color=None, label=None, stacked=False, *, data=None, **kwargs)

```python
# 조금 더 자세히 알아보자.


# 데이터 교점 좌표 8개
x = [0,1,2,3,4,5,6,7]


# 사이즈의 증가율을 컨트롤하는 예시 3가지 2^n, n^2, n
# for loop를 list 안에 내포하였다. (리스트 내포)
# 이 데이터의 경우, range(len(x)) ➡ range(8) 이 됨. (range 객체 리턴, for loop가능.)
# 각 예시에 *10 을 하면, 변화율의 차이가 더 잘 보임
s_exp = [2**n for n in range(len(x))]  # 0~7 까지 n에 대해 2^n 값들의 list
s_square = [n**2 for n in range(len(x))] # 0~7 까지 n에 대해 n^2 값들의 list
s_linear = [n for n in range(len(x))] # 0~7 까지 n에 대해 n 값들의 list

# y값의 좌표는 1,0,1로 통일, 각 size들 적용.
plt.scatter(x,[1]*len(x),s=s_exp, label='$s=2^n$')
plt.scatter(x,[0]*len(x),s=s_square, label='$s=n^2$')
plt.scatter(x,[-1]*len(x),s=s_linear, label='$s=n$')

# axes 사이즈는 넉넉하게
plt.ylim(-2,2)

# 범례의 위치 조정.
# 참고로 figure는 좌하단(0,0) ~ 우하단(1,1) 위치값으로 되어 있다.
# 범례의 위치를 (1.2,0.5)로 둔 것은, figure에서 x축 양의 방향으로 살짝 벗어나게 + y축은 가운데에 위치시킨 것이다.
plt.legend(loc='center', bbox_to_anchor=(1.2, 0.5), labelspacing=3)
plt.show()
```

```python
# 👀참고 : plot()의 markersize 와 scatter의 size 사이의 관계
# size가 더 본질이다. size는 marker가 그려질 정방형 영역(square area)의 한 변의 길이이다.
# markersize = size**2 이다.

# 그 정방형 영역을, Data Point 라고 하고, 1point = 1pixels 로 봐도 무방. (72dpi 에서는 1pt = 1px임. 더 깊은건 여기서 X)
# 더 알고싶은 사람 = https://spoqa.github.io/2012/07/06/pixel-and-point.html

plt.plot([1], [1], 'o', markersize=20, c='r')
plt.scatter([2], [1], size=20**2, c='b')

plt.xlim(0,3)
```

### **4.Pie 차트를 만들어보자.**
#### 1) 기본 조건  
#### 아래 링크의 세계 반도체 시장 점유율 그래프를 구현하되,
#### https://www.notebookcheck.net/TSMC-has-the-largest-market-share-in-the-global-semiconductor-manufacturing-industry-Counterpoints-Research.518137.0.html

#### 2) Shadow 효과를 줄 것.

#### 3) KOREA와 JAPAN 에 explode 효과를 주되,KOREA를 좀더 explode 시킬 것.

#### 4) autopct 기능을 사용할 것

#### 5) 컬러 팔레트는 쓰지 말 것.


```python
import seaborn as sns
import matplotlib.pyplot as plt

semiconductor = ['Korea', 'China', 'Taiwan', 'Japan', 'USA', 'EU', 'etc']
rate = [24.4, 28.8, 24.3, 7.6, 7.4, 3.2, 4.3]

ex_rate = (0.1, 0, 0, 0.07, 0, 0, 0) # 각 wedge들의 explode 돌출율을 위한 튜플값

palette = sns.color_palette('pastel',len(semiconductor)) # pastel 팔레트에서 brand 수 만큼 색상 가져오기

plt.pie(rate, explode = ex_rate, labels=semiconductor, colors=palette, shadow=True, autopct="%0.1f%%")
```

### **5.Scatter 차트를 만들어보자.**
#### 1) [1,10) 범위의 **정수** 난수값을 마음대로 갯수를 정하여 x,y 좌표 데이터로 쓰자. (범위는 꼭1~10 이어야 함.)
#### 2) 마커의 사이즈는 y값이 올라갈 수록, 2^y가 되어 커지게 만들자.
#### 3) 각 마커에 랜덤 컬러를 적용하고 투명도는 0.5로 하자.  
#### 4) colormap을 spring으로 하자.
#### 5) 필요한 경우, axes는 넉넉하게 키워주자.


```python
# 마커 사이즈 변화에 규칙성을 부여해보자

# 마커 100개(교점100개) 생성
x = np.random.randint(1,10,100)
y = np.random.randint(1,10,100)

# 마커 컬러 : 교점이 100개 발생했기 때문에, 마커 컬러도 100개에 대해 지정해줘야 한다.
colors = np.random.rand(100) #(100개의 난수 시퀀스로 랜덤 컬러 지정)

# 마커 크기 : 교점이 100개 발생했기 때문에, 마커 사이즈도 100개에 대해 지정해줘야 한다.
# y값이 커지느냐 작아지느냐가 마커 사이즈 변동의 원인이므로, y축방향에 따라 사이즈 변화가 일어난다.
area = 2**y


plt.scatter(x, y, s=area, c=colors, alpha=0.5) #alpha = 알파채널, 투명도(0~1)
plt.show()
```
