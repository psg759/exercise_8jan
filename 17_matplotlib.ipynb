##‚úÖ **Ïä§ÌÉÄÏùº ÏïΩÏñ¥ (Style Abbrebiation)**

```python
# Ïä§ÌÉÄÏùº Ï¢ÖÌï©
plt.plot([1,2,3,4,5],[2,4,6,8,10],
         color="dodgerblue",
         linewidth=2,
         linestyle="-.",
         marker="o",
         ms=7,
         mec="violet",
         mew=2,
         mfc="mediumslateblue")
plt.show()
```

##‚úÖ **title, xlabel, ylabel**
## Í∑∏ÎûòÌîÑ ÏûêÏ≤¥ Ïù¥Î¶ÑÍ≥º, xÏ∂ï, yÏ∂ï Ïù¥Î¶Ñ ÏßÄÏ†ï

##‚úÖ **xlim, ylim**
## axes ÏòÅÏó≠ÏùÑ Î≥ÄÍ≤ΩÌïòÎäî Í≤É.
## Î≥¥ÌÜµÏùÄ axes ÏòÅÏó≠Ïù¥ Ïù∏ÏûêÎ°ú Ìà¨ÏûÖÎêú Í∞íÏóê Ï¢ÖÏÜçÎêòÎÇò, Î≥ÑÎèÑÎ°ú Î≥ÄÍ≤Ω Ïã∂ÏùÑ Îïå ÏÇ¨Ïö©.

```python
# xÎ≤îÏúÑ, yÎ≤îÏúÑ Î™®Îëê 100 Ïù∏ Í∑∏ÎûòÌîÑÏù¥ÏßÄÎßå,
# axes ÏòÅÏó≠ÏùÑ Î≥ÄÍ≤ΩÌïòÏó¨, xÎäî 75ÍπåÏßÄ, yÎäî 200ÍπåÏßÄ axesÎ•º ÌëúÏãú

x = np.arange(0,100)
y = np.arange(0,100)

plt.plot(x,y)

plt.xlim(0,75)
plt.ylim(0,200)

plt.show()
```

##‚úÖ **legend**
## Î≤îÎ°Ä ÏÑ§Ï†ï.
## plt.plot(label='Î≤îÎ°ÄÎ™Ö')
## plt.legend(loc='Î≤îÎ°ÄÏúÑÏπò')
### best(ÏûêÎèô) , center(Ï§ëÏïô)

##‚úÖ **tick**
## ÎààÍ∏à ÏÑ§Ï†ï. ÎààÍ∏àÏùò **Í∞ÑÍ≤©**Í≥º
## **ÌëúÍ∏∞Î¨∏ÏûêÏùò Í∞ÅÎèÑ**Î•º ÏàòÏ†ïÌï† ÏàòÏûàÏùå

## matplotlib.pyplot.xticks(ticks=None, labels=None, *, minor=False, **kwargs)
## ‚ñ™Ô∏è tics : Í∞Å ÎààÍ∏àÎì§Ïùò ÏúÑÏπò(position)Î•º Îã¥ÏùÄ array-likeÌïú Í≤É
## ‚ñ™Ô∏è labels : Í∞Å ÎààÍ∏àÎì§Ïùò Ïù¥Î¶Ñ(label)ÏùÑ Îã¥ÏùÄ array-likeÌïú Í≤É

<br>

## plt.xticks(ticks = np.arange(start,stop,step)) üëâ xÏ∂ï ÎààÍ∏à Í∞ÑÍ≤©
## plt.yticks(ticks = np.arange(start,stop,step)) üëâ yÏ∂ï ÎààÍ∏à Í∞ÑÍ≤©
## plt.xticks(rotation='angle') üëâ xÎààÍ∏à Î¨∏ÏûêÏó¥ ÌöåÏ†Ñ
## plt.yticks(rotation='angle') üëâ yÎààÍ∏à Î¨∏ÏûêÏó¥ ÌöåÏ†Ñ

```python
# Í∑∏ÎÉ• plotÏùÑ ÎßåÎì§Î©¥ default Í∞íÏùÄ ÏïÑÎûòÏôÄ Í∞ôÏùå.
# plotÏùò Î≤îÏúÑÎäî x(0~9),y(0~9)Ïù∏Îç∞ ÎààÍ∏àÏùÄ 0,2,4,6,8

x = np.arange(0,10)
y = np.arange(0,10)

plt.plot(x,y)

plt.show()
```

```python
# ticksÎ•º plotÏùò x,yÍ∞íÍ≥º ÎèôÏùºÌïòÍ≤å Ìï¥Ï£ºÎ©¥,
# Ï§ëÍ∞ÑÏùÑ Í±¥ÎÑàÎõ∞ÏßÄ ÏïäÏùÄ, plot : tics = 1:1 ÎåÄÏùëÏùò ticsÍ∞Ä ÏÉùÍπÄ

x = np.arange(0,10)
y = np.arange(0,10)

plt.plot(x,y)

plt.xticks(ticks = x)
plt.yticks(ticks = y)

plt.show()
```

```python
# plotÏù¥ Ï≤®ÏòàÌï¥ÏßÄÎ©¥?

x = np.arange(0,100)
y = np.arange(0,100)

plt.plot(x,y)

plt.xticks(ticks = np.arange(0,100))
plt.yticks(ticks = np.arange(0,100))

plt.show()
```

```python
# plotÏù¥ Ï≤®ÏòàÌï¥ÏßÄÎ©¥ ticsÎ•º Î™®Îëê ÌëúÍ∏∞ÌïòÍ∏∞Î≥¥Îã§, Í∞ÑÍ≤©ÏùÑ ÎëêÎäîÍ≤å ÎÇòÏùå.
# Í∑∏ÎûòÏÑú ticsÎäî plotÏùò (x,y) Í∞úÏàòÏôÄ 1:1 ÎåÄÏùëÏù¥ ÏïÑÎãå, ÏûëÏùÄ Í∞íÏùÑ Í∞ÄÏßà Ïàò ÏûàÏùå.

x = np.arange(0,100)
y = np.arange(0,100)

plt.plot(x,y)

plt.xticks(ticks = np.arange(0,100,5)) #xÏ∂ïÏùÄ 5Í∞ÑÍ≤© ÎààÍ∏à
plt.yticks(ticks = np.arange(0,100,10)) #yÏ∂ïÏùÄ 10Í∞ÑÍ≤© ÎààÍ∏à

plt.show()
```

## ‚ñ™Ô∏è explode : Í∞Å ÌååÏù¥Îì§Ïùò ÎèåÏ∂úÏ†ïÎèÑÎ•º ÌäúÌîåÎ°ú Î≥ÑÎèÑ Ï†ïÏùòÌïòÏó¨ Ï†ÅÏö©Ìï®.
